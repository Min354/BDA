```python
print("sum".ljust(12), str(
    np.sum(tt['fare'])
    ), sep=' : ')

print("sum(not np)".ljust(12), str(
    sum(tt['fare'])
    ), sep=' : ')

print("median".ljust(12), str(
    np.median(tt['fare'])
    ), sep=' : ')

print("var".ljust(12), str(
    np.var(tt['fare'])
    ), sep=' : ')
```
