### 기초 통계치
- [''].mean() / ''.mean()
- ''.sum()
- ''.count()
- ''.median()
- 기초 통계치 간 기본적인 산수 연산 가능
---
- groupby('group columns')['column'].f()


```python
print("sum".ljust(12), str(
    tt.age.sum()
    ), sep=' : ')

print("sum[]".ljust(12), str(
    tt['age'].sum()
    ), sep=' : ')

print("median".ljust(12), str(
    tt.age.median()
    ), sep=' : ')

print("count".ljust(12), str(
    tt.age.count()
    ), sep=' : ')
```
