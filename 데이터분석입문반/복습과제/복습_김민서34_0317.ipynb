### 파생변수 만들기
- 기존 변수 이용(산술 연산, 기존 변수 합치기 등)

```python
tt['fare1000'] = 0
tt['fare1000'] = tt.fare * 1000
tt
```

```python
fMean = tt.query(''' sex=="female" ''')['age'].mean()
mMean = tt.query(''' sex=="male" ''')['age'].mean()
tt['femaleAgeMean'] = 0
tt['maleAgeMean'] = 0
```

```python
for i in range(len(tt)):
    if tt['sex'].loc[i] == 'female':
        tt['femaleAgeMean'].loc[i] = fMean
    elif tt['sex'].loc[i] == 'male':
        tt['maleAgeMean'].loc[i] = mMean
tt[['femaleAgeMean', 'maleAgeMean']]
```

### assign(), lambda
1. assign(): 파생변수 할당 함수
- assign(변수명 = np.where(조건, (True:)값, (False:)값))
- np.where(조건): 조건에 해당하는 인덱스 반환
2. lambda

```python
import numpy as np
```

```python
tt.assign(v1 = np.where(tt['fare']>50, '1','0'))
```

```python
tt  #v1 적용안됨
```

```python
np.where(tt['age']>70)
```

```python
ttNew = tt.assign(age_old = np.where(tt['age']>30, 'old', 'young'))
ttNew 
```

---

```python
tt.assign(v1 = lambda x: x['fare']*1000)
```
