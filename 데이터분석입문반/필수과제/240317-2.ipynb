```python
import seaborn as sns
tt = sns.load_dataset('titanic')

tt['fare_mean'] = 0

for i in range(len(tt)):
    if(tt['sex'].loc[i] == 'female'):
        if(tt['age'].loc[i] < 10):
            tt['fare_mean'].loc[i] = tt.query(''' sex=="female" & age<10 ''')['fare'].mean()
        elif(tt['age'].loc[i] < 20):
            tt['fare_mean'].loc[i] = tt.query(''' sex=="female" & 10<=age<20 ''')['fare'].mean()
        elif(tt['age'].loc[i] < 30):
            tt['fare_mean'].loc[i] = tt.query(''' sex=="female" & 20<=age<30 ''')['fare'].mean()
        elif(tt['age'].loc[i] < 40):
            tt['fare_mean'].loc[i] = tt.query(''' sex=="female" & 30<=age<40 ''')['fare'].mean()
        else:
            tt['fare_mean'].loc[i] =  tt.query(''' sex=="female" & age>=40 ''')['fare'].mean()
        
    elif(tt['sex'].loc[i] == 'male'):
        if(tt['age'].loc[i] < 10):
            tt['fare_mean'].loc[i] = tt.query(''' sex=="female" & age<10 ''')['fare'].mean()
        elif(tt['age'].loc[i] < 20):
            tt['fare_mean'].loc[i] = tt.query(''' sex=="female" & 10<=age<20 ''')['fare'].mean()
        elif(tt['age'].loc[i] < 30):
            tt['fare_mean'].loc[i] = tt.query(''' sex=="female" & 20<=age<30 ''')['fare'].mean()
        elif(tt['age'].loc[i] < 40):
            tt['fare_mean'].loc[i] = tt.query(''' sex=="female" & 30<=age<40 ''')['fare'].mean()
        else:
            tt['fare_mean'].loc[i] =  tt.query(''' sex=="female" & age>=40 ''')['fare'].mean()
```
