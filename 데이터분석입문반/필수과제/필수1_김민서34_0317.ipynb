```python
#2. survived & male & age<=10
tt.query(''' survived==1 & sex=="male" & age<=10 ''')
```

```python
#3. !survived | age<=10 | alone=="yes"
tt.query(''' survived==0 | age<=10 | alone=="yes"  ''')
```

```python
#4. (!survived & age<=10) | alone
tt.query(''' (survived==0 & age<=10) | alone==True  ''')
```
